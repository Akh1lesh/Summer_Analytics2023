```python
a = []
for i in range(len(x)):
  a.append(list(x[i]))
```

```python
for i in range(len(a)):
  a[i].append(1)
```

```python
a = np.array(a)
```

```python
a_new = []
for i in range(len(a)):
  a_new.append(a[i])
a_new
```

```python
a_transpose = np.transpose(a_new)
```

```python
pseudo_a = np.dot((np.linalg.inv(np.dot(a_transpose, a_new))),np.transpose(a_new))
```

```python
df.shape
```

```python
x.shape
```

```python
pseudo_a.shape
```

```python
y_train = np.array(y_train)
y_train.shape
```

```python
parameters = np.dot(pseudo_a ,y_train)
```

```python
parameters.shape
```

```python
a_new[0].shape
```

```python
y_pred = []
for i in range(len(a)):
  y_pred.append(np.dot(np.transpose(parameters),a_new[i]))
```

```python
threshold = np.median(y_pred)
threshold
```

```python
for i in range(len(y_pred)):
  if(y_pred[i] < threshold):
    y_pred[i] = 0
  else:
    y_pred[i] = 1
```

```python
y_pred
```

```python
#Accuracy of the model
count = 0
for i in range(len(y_pred)):
  if(y_pred[i] == y_train[i]):
    count = count + 1

accuracy = (count/len(y_pred)) * 100
print(accuracy)
```
